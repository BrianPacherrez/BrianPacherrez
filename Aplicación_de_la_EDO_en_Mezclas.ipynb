#**¬øQU√â SON LAS MEZCLAS?**
Una mescla es la uni√≥n f√≠sica de dos o mas sustancias que cumplen las siguientes condiciones: 
*   cada una de las sustancias componentes conserva sus propiedades. 
*   Las sustancias componentes son separables por medios f√≠sicos 
*   Las sustancias componentes pueden intervenir en cualquier proporci√≥n.
*   En su formaci√≥n las mesclas no presentan manifestaciones energ√©ticas.
Dicho de otro modo, una mescla es un agregado de dos o mas sustancias entre las que no existe acci√≥n qu√≠mica alguna.

separacion-de-mezclas-e1579277187397.jpg


Definamos la concentraci√≥n de una sustancia como:
Concentraci√≥n = Cantidad de sustancia/Volumen total
Cuando tenemos un recipiente conteniendo una mezcla homog√©nea; el cual tiene una entrada y una salida; entonces:
En un instante cualquiera una sustancia presente en la mezcla se definir√° como:
# $$\frac{dQ(t)}{dt}=Qe - Qs$$

Donde:
*   Q(t) = cantidad de sustancia.
*   Qe = cantidad de sustancia de entrada.
*   Qs = cantidad de sustancia de salida.

Adem√°s, sabemos que:
*   Qe = Ve.Ce.
*   Qs = Vs.Cs.

Donde:
*   Ve = Volumen entrante.
*   Vs = Volumen de salida.
*   Ce = Concentraci√≥n de entrada.
*   Cs = Concentraci√≥n de salida.

El volumen en un tiempo cualquiera ser√°:

$$ V(t)=V0 + (Ve - Vs)t $$

Donde:
*  V (t) = cantidad de sustancia.
*  Ve = cantidad de sustancia de entrada.
*  Vs = cantidad de sustancia de salida.

Entonces la concentraci√≥n de la sustancia en el recipiente ser√°:

$$C(t)=\frac{Q(t)}{V(t)}$$
