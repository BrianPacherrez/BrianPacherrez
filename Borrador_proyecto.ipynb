$$\frac{dx}{dt}=-\frac{5x}{1000+t}$$

```python
from sympy import *  # Libreria CAS
init_printing()      # para ver los resultados en simbologia Latex
```

```python
t=symbols("t")   # defino las variables simbolicas "x"
x=Function("x")  # defino la función "y"

ed = Eq(x(t).diff(),-5*x(t)/(1000+t))
ed
```

```python
sol=dsolve(ed,x(t))
sol
```

```python
simplify(sol)
```

```python
x=symbols("x")   # defino las variables simbolicas "x"
t=Function("t")  # defino la función "y"

ed1 = Eq(t(x).diff(),-(5*x/1000+t(x)))
ed1
```

```python
sol1=dsolve(ed1,t(x))
sol1
```

```python
simplify(sol1)
```
