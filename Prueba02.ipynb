$Q=500 gal \\ C_{1}=2\frac{lb}{gal} \\ V_{1}=5\frac{gal}{min} \\ V_{2}=10\frac{gal}{min} \\  x(t)=1 lb de sal$






## Solución:

$$\frac{dx}{dt}+\frac{V_{2}}{Q+(V_{1}-V_{2})t}x=V_{1}.C_{1}$$

$$\frac{dx}{dt}+\frac{10}{500+(5-10)t}x=5.2$$

$$\frac{dx}{dt}+\frac{10}{500-5t}x=10$$

$$\frac{dx}{dt}+\frac{5.2}{5(100-t)}x=10$$

$$\frac{dx}{dt}+\frac{2}{(100-t)}x=10$$

$$\frac{2}{(100-t)} \Rightarrow p(t)$$
