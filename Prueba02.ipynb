## $$\frac{dx}{dt}+\frac{2}{(100-t)}x=10$$

```python
t=symbols("t")   # defino las variables simbolicas "x"
x=Function("x")  # defino la función "y"

ed = Eq(x(t).diff()+(2/(100-t))*x(t),10)
ed
```

```python
dsolve(ed,x(t))
```
