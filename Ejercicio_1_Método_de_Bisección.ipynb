## $f(x)=x.\cos(x+1)$

### De la siguiente gráfica de la función, escogemos los intervalos $[a,b]$.

```python
x=symbols("x")
plot(x*cos(x+1),(x,-1.5,14.5),ylim=(-12,12),line_color="red")
```

### Desarrollo: 
Ingresamos los parámetros (la función, los extremos de la función $[a,b]$ y el error de aproximación (tolerancia)).

\\
*   Función: $ \,\ x.\cos(x+1)$
*   Extremos de la función: $ \,\ [10.1,13.5]$
*   Error de aproximación: $ \,\ 0.000000000001$

```python
fx = lambda x: x*np.cos(x+1) 
a = 10.1
b = 13.5
tolerancia = 0.000000000001
```

### Procedimiento

```python
# Creamos un array y según las condiciones, se hallarán los diferentes resultados 
# Después serán mostrados en la tabla
tabla = []
tramo = b-a

fa = fx(a)
fb = fx(b)
i = 1
while (tramo>tolerancia):
    c = (a+b)/2
    fc = fx(c)
    tabla.append([i,a,c,b,fa,fc,fb,tramo])
    i = i + 1
                 
    cambia = np.sign(fa)*np.sign(fc)
    if (cambia<0):
        b = c
        fb = fc
    else:
        a=c
        fa = fc
    tramo = b-a
c = (a+b)/2
fc = fx(c)
tabla.append([i,a,c,b,fa,fc,fb,tramo])
tabla = np.array(tabla)

raiz = c
```

### Salida

```python
np.set_printoptions(precision = 4)
print('[i,       a,        c,        b,      f(a),     f(c),     f(b),     tramo ]')
# Colocamos el título o simbolo de los valores de los resultados, para mostrarlos junto con la tabla

# Le damos un formato a la tabla, ordenandola y para visualizar valores de 5 decimales
n=len(tabla)
for i in range(0,n,1):
    unafila = tabla[i]
    formato = ' {:<2.0f}'+' '+(len(unafila)-1)*'{:>10.5f}'
    unafila = formato.format(*unafila)
    print(unafila)
    
print('raíz: ',raiz)  # Mostramos el valor de la raíz de la función
```

### Gráfica usando Matplotlib

```python
xi = tabla[:,2]
yi = tabla[:,5]

# ordenamos los puntos para la gráfica
orden = np.argsort(xi)
xi = xi[orden]
yi = yi[orden]

plt.plot(xi,yi)
plt.plot(xi,yi,'o')
plt.axhline(0, color="black")

plt.xlabel('x') # Mostramos el eje x
plt.ylabel('y') # Mostramos el eje y
plt.title('Bisección en f(x)') # Mostramos el título de la gráfica
plt.grid()
plt.show()
```
